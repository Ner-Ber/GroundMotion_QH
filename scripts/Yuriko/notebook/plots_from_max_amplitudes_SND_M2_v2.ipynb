```python
import numpy as np
import matplotlib.pyplot as plt

from ground_motion_qh.plotting import (
    scatter_post_ratio_vs_apre,
    scatter_pre_vs_post,
    exceedance_plot,
    hist_apre_apost,
    box_plot_shake_ratio_vs_apre
)
```

```python

# # NOISE
# a_post_max = np.load("../../results/a_forecast_max.npy")
# a_pre_max = np.load("../../results/a_trigger_max.npy")

#outdir='/home/yiwasak2/QuakeHack/GroundMotion_QH/data/raw_data/ANZA_SND_1994_M2_Picks'
outdir='/home/yiwasak2/QuakeHack/GroundMotion_QH/data/raw_data/ANZA_SND_1994_M2.0_Picks'

# USED FOR THURSDAY PRESENTATION
a_post_max = np.load("%s/a_max_plus.npy" % outdir)
a_pre_max = np.load("%s/a_max_minus.npy" % outdir)

a_post_max = np.array([a_post_max[i] for i in np.arange(0, len(a_post_max), 2)])
a_pre_max = np.array([a_pre_max[i] for i in np.arange(0, len(a_pre_max), 2)])

# # YURIKO, USING OLD PROCESSING FUNCTION
# a_post_max = np.load("../../results/Yuriko/a_forecast_max_yi.npy")
# a_pre_max = np.load("../../results/Yuriko/a_trigger_max_yi.npy")
```

```python
fig, ax = scatter_post_ratio_vs_apre(
    a_pre_max,
    a_post_max,
)
plt.savefig("./figures/scatter_basic.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
fig, ax = scatter_pre_vs_post(
    a_pre_max,
    a_post_max,
)
plt.savefig("./figures/contours.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
fig, ax = exceedance_plot(
    a_pre_max,
    a_post_max,
    n_roll=100
)
plt.savefig("./figures/exceedance.png", dpi=300, bbox_inches="tight")
plt.show()
```
